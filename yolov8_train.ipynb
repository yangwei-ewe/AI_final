```python
from ultralytics import YOLO
from os import system
import cv2
```

```python
def getAnswer(resault:list)->list[str]:
    rtn=[]
    for it in resault:
        rtn.append("".join([chr(ord("A")+int(it[-1])) for it in sorted(it.boxes.data,key=lambda box: box[0])]))
    return rtn
```

```python
system("rm -r runs/detect/train*")
model = YOLO("yolov8n.pt")
model.train(data="train.yaml", mode="detect", epochs=1000, imgsz=480)
```

```python
system("rm -r runs/detect/predict*")
modelTrain = YOLO("./runs/detect/train/weights/best.pt")
trainResault=modelTrain.predict(
    source="./datasets/images/val",
    mode="predict",
    save=True
)
```

```python
print((101-len([it for it in getAnswer(trainResault) if len(it)!=4]))/101)
```
